```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1.Line Chart (trend)
plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Indraniel')
plt.show()
```

```python
#Advanced Line Chart
plt.figure(figsize=(7,4))
plt.plot(df['Days'], df['Study_Hours'], marker='o', color='blue',label='Study Hours')
plt.plot(df['Days'], df['Marks'], marker='s', color='red',label='Marks')
plt.title('Indraniel')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#2. Bar Chart (comparision)
plt.bar(df['Days'], df['Marks'], color='skyblue')
plt.title('Indraniel')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Days'], df['Marks'], color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title('Indraniel')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(len(df['Days'])) #positions fors bars
width=0.35 #width of each bar
plt.bar(x-width/2,df['Study_Hours'],width,label='Study Hours')
plt.bar(x+width/2,df['Marks'],width,label='Marks')
plt.xticks(x,df['Days'])
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Indraniel')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
marks=[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
mean_value=np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=5,color='skyblue',edgecolor='black')
plt.axvline(mean_value,color='red',linestyle='dashed',linewidth=2,label='Mean')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Indraniel')
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
#3. Histogram (Distribution)
#Shows how values are spread
data={'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=3)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Indraniel')
plt.show()
```

```python
#4. Scatter plot (Relationship)
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
plt.scatter(df['Study_Hours'],df['Marks'])
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Indraniel')
plt.show()
```

```python
#conditional Scatter Plot
#Example Category
df['Result'] = ['Fail','Pass','Fail','Pass','Pass']
colors={'Fail':'red','Pass':'green'}
# Map the 'Result' column to actual colors
scatter_colors = [colors[result] for result in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=scatter_colors)
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Indraniel')
plt.show()
```

```python
#Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics','Food',
                 'Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
print(df)
```
